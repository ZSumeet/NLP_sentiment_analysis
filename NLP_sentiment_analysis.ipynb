```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://www.yelp.com/biz/tesla-san-francisco?osq=Tesla-Dealership")
```

```python
r.status_code
```

```python
r.text
```

```python
soup = BeautifulSoup(r.text, "html.parser")
div =  soup.findAll(class_="review-content")
```

```python
print(len(div))
```

```python
review = []
for divs in div:
    review.append(divs.find('p').text)
```

```python
review[0]
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.array(review), columns = ["review"])
```

```python
df.head()
```

```python
len(df['review'])
```

```python
df['word_count'] = df['review'].apply(lambda x: len(x.split()))
```

```python
df.head()
```

```python
df['character_count'] = df['review'].apply(lambda x: len(x))
```

```python
df.head()
```

```python
def average_word(x):
    words = x.split()
    return sum(len(words) for word in words)/ len(words)
  
```

```python
df['average_word_length'] = df['review'].apply(lambda x: average_word(x) )
```

```python
from nltk.corpus import stopwords
```

```python
stopword = stopwords.words('english')
```

```python
df['stopword_count'] = df['review'].apply(lambda x: len([word for word in x.split() if word.lower() in stopword]))
```

```python
df['stopword_rate'] = df['stopword_count']/df['word_count']
```

```python
df.sort_values(by = "stopword_rate")
```

```python
df['review']
```

```python
df['lower_case'] = df['review'].apply(lambda x: " ".join(word.lower() for word in x.split()))
```

```python
df.head()
```

```python
df['punctuation'] = df['lower_case'].str.replace('[^\w\s]', '')
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
df['stopword']  =df['punctuation'].apply(lambda x: " ".join(word for word in x.split() if word not in stop_words))
```

```python
df.head()
```

```python
pd.Series(" ".join(df['stopword']).split()).value_counts()[:30]
```

```python
other_stopword =['get', 'told', 'would', 'week','us', 'right', 'even', 'also', 'asked']
```

```python
df['clean_data'] = df['stopword'].apply(lambda x: " ".join( word for word in x.split() if word not in other_stopword))
```

```python
df.head()
```

```python
df['clean_review_word_count'] = df['clean_data'].apply(lambda x: len(x.split()))
```

```python
import textblob, nltk
```

```python
nltk.download('wordnet')
```

```python
from textblob import Word
```

```python
df['lematized']  = df['clean_data'].apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
```

```python
df.head()
```

```python
from textblob import TextBlob
```

```python
df['polarity'] = df['lematized'].apply(lambda x: TextBlob(x).sentiment[0])
```

```python
df['subjectivity'] = df['lematized'].apply(lambda x: TextBlob(x).sentiment[1])
```
